```python
# Meta Data of Dates and Ball Numbers (Stored Locally and Retrieved Online) - http://www.lottoreport.com/ticketcomparison.htm

metadf = pd.read_excel("megametadata.xlsx")
df=pd.read_csv('masterlottery.csv')
metadf.index=df.index
df["megaplier"]=metadf["megaplier"]
df["date"]=metadf["date"]
df["ball1"]=metadf["ball1"]
df["ball2"]=metadf["ball2"]
df["ball3"]=metadf["ball3"]
df["ball4"]=metadf["ball4"]
df["ball5"]=metadf["ball5"]
df["megaball"]=metadf["megaball"]
df["totaltickets"]=metadf["totaltickets"]
df["jackpot"]=metadf["jackpot"]
test=df.copy()
numbercolumns=['5+1', '5+0', '4+1', '4+0', '3+1', '3+0','2+1', '1+1', '0+1', '5+0+M', '4+1+M', '4+0+M', '3+1+M', '3+0+M','2+1+M', '1+1+M', '0+1+M']
dfnumeric = pd.read_csv("masterLottery.csv", index_col=0, dtype=np.float64,usecols=numbercolumns,thousands=',')
dfnumeric.index = df.index

test=df.copy()

## Calculated Odds Of Continued Success from 07/27/2018 (at 40 million) until 10/23/2018 (at 1.537 billion) -

test['odds'] = test.apply(
    lambda row: (302575349/302575350)**row['totaltickets']
    ,axis=1)

i=0
record3=[]
placer=1
for x in test["odds"]:
    
    placer*=x
    i+=1
    record3.append(placer)
    if i == 25:
        break
        
formatrecord = [ '%.3f' % elem for elem in record3 ]
# Perform Calculations to get T scores (accordingly to the T Test equation above)

test["expectedmegawinners"]=round(test["totaltickets"]*1/25,0)
test["actualmegawinners"]=dfnumeric[["4+1","3+1","2+1","1+1","0+1"]].sum(axis=1)
test["discrepancy"]=test["expectedmegawinners"]-test["actualmegawinners"]
test["expectedstd"]=(test["totaltickets"]*1/25*(24/25))**(1/2)
test["actualmean"]=test["actualmegawinners"]/test["totaltickets"]
test["actualstd"]=(test["totaltickets"]*test["actualmean"]*(1-test["actualmean"]))**(1/2)
test["tscores"]=round((test["actualmegawinners"]-test["expectedmegawinners"])/(test["actualstd"]/(test["totaltickets"])**(1/2)),0)

# Create T Tables DF

ttables=test[["date","megaball","tscores","actualmean","totaltickets","actualstd","actualmegawinners"]]
# For Loop to Collect Standard Deviations for Every MegaBall

Deviations=pd.DataFrame(columns=["std","count"],index=range(1,26))
for x in range(0,25):
    Deviations.iloc[x,1]=ttables[ttables.megaball==x+1].count()[2]
    Deviations.iloc[x,0]=ttables[ttables.megaball==x+1].std()[2]



# Create Box Plot Data (List of Lists for every megaball mean)

def hello(x):
    return(np.array(pd.DataFrame(ttables[ttables["megaball"]==x]["actualmean"])))

boxlist=[]
for y in range(1,25):
    boxlist.append(list(hello(y)))
groupedtable=ttables.groupby(['megaball'])[('actualmean')].agg(['mean', 'count'])
popularballs=groupedtable.sort_index().sort_values("mean",ascending=False)
```

```python
import math
import zipfile
from io import BytesIO
from requests import get
```

```python
def retrievezip(link,filenum=1):
    request = get(link)
    zf = zipfile.ZipFile(BytesIO(request.content)) 
    files = zf.namelist()
    return(pd.read_csv(zf.open(files[filenum]),skiprows=4, delim_whitespace=False, header=0))
```

```python
df1 = retrievezip("http://api.worldbank.org/v2/en/indicator/ST.INT.ARVL?downloadformat=csv", 1)
df2 = retrievezip('http://api.worldbank.org/v2/en/indicator/NY.GDP.MKTP.CD?downloadformat=csv', 1)
df3 = retrievezip('http://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv', 1)
df4 = retrievezip('http://api.worldbank.org/v2/en/indicator/ST.INT.RCPT.CD?downloadformat=csv',1)
```

```python
# Function to select years in databases

def keepyears(dataframe, startyear, endyear, yearstart=3):
    
    sliceleft=dataframe.drop(dataframe.columns[yearstart:dataframe.columns.get_loc(startyear)], axis=1)
    sliceright=sliceleft.drop(sliceleft.columns[sliceleft.columns.get_loc(endyear)+1:], axis=1)
    return(sliceright)

start="2000"
end="2016"

df1=keepyears(df1, start, end)
df2=keepyears(df2, start, end)
df3=keepyears(df3, start, end)
df4=keepyears(df4, start, end)
```

```python
# Naming of Dataframes

df1.name=df1.iloc[1,2]
df2.name=df2.iloc[1,2]
df3.name=df3.iloc[1,2]
df4.name=df4.iloc[1,2]
```

```python
# Keep Rows of 6 East African countries and Name Index

countryindex=["Burundi", "Ethiopia", "Kenya", "Rwanda", "Tanzania", "Uganda"]

def selectcountries(dataframe, countries):
    dataframe.index=dataframe["Country Name"]
    return(dataframe[dataframe["Country Name"].isin(countries)])
    
df1=selectcountries(df1, countryindex)
df2=selectcountries(df2, countryindex)
df3=selectcountries(df3, countryindex)
df4=selectcountries(df4, countryindex)
```

```python
def plotdataframe2(*dataframes):
    
    plotsize=0
    for dataframe in dataframes:
        plotsize+=1
    
    plt.figure(figsize=(10, plotsize*3))
    
    plotcounts=0
    for dataframe in dataframes:
        plotcounts+=1
        
        plt.subplot(plotsize, 1, plotcounts)
        plt.title("Graph "+str(plotcounts)+" - "+dataframe.name)
        for country in countryindex:
            plt.plot(dataframe.loc[country,start:end])
        plt.legend(bbox_to_anchor=(1.02, 1.0), loc=2, borderaxespad=0)

    plt.tight_layout()
    plt.show()
               
plotdataframe2(df1,df2,df3,df4)
```
