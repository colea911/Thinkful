The lucky single jackpot winner won 1.537 billion dollars, which is the 2nd largest amount ever won in a lottery in the US.  This is an almost incomprehensible amount. ## Background - The Rules of the Game

Before delving into probabilities and economics, it is worth refreshing our minds on the rules of this game.  

#### Cycles starting at 40 million

The Mega Millions is a lottery that goes through cycles of building up its massive jackpot amount.  At the beginning of the cycle, the lottery's jackpot starts at 40 million.  Every Tuesday and Friday, a series of 5 balls, numbered 1 to 70, and a special "Mega Ball", numbered 1 to 25, is randomly drawn from a machine (without replacement).  These numbers are broadcasted and anyone who can match all the numbers wins the jackpot.   

#### Cycle of increasing jackpot and players

If there are no winners, than almost all the money collected from ticket sales are returned into the jackpot for the sebsequent cycle.  As you can imagine, if nobody wins for an extended period of time, the jackpot will continue growing.  As the jackpot continues to grow, so do ticket sales and the chances of someone randomly having the lucky jackpot number increases.  

#### Other Rules

Beyond the jackspot, there are other rules and prizes to encourage players.  There are prizes for matching any one of the balls, and the winning go up for the higher level of balls matched.  Furthermore there is an optional multiplier number that can be purchased for an additional dollar which will any winnings (exclusing the jackpot) by a randomly selected multiplier between 1 and 5.

#### Ticket Cost and Recent Rules

To play the Mega Millions one must pay for a 2 dollar ticket and select there numbers (no repeats!).  In 2017 October 27, the Mega Millions changed their rules to increase the price of a ticket from 1 to 2 dollars.  Furthermore they made it so the balls went to a higher count (from 50 to 70).  These rules were likely put into place to make it so that the jackpot would reach a larger amount.  By increasing the ticket price, more money will flow into the jackpot.  By increasing the numbers of balls, it is less likely in any given cycle to have a jackpot winner.

#### Random Drawing

The balls are randomly drawn from a machine.  Presumably, there is no bias.  To get a flavor for how this is done, take a brief look at the recently winning draw below

## What Are the Chances of Winning?

This question is fairly simple, and is in fact quite transparent.  The chances are 1 in 302,575,350.

It is worth reminding ourselves of the math in combinatorics.  There is a pool of 70 balls for the first 5 choices of balls, and a total of 25 balls for the last 6th choice.  Therefore we look at the odds of getting all the chosen balls one by one.  The equation below produces the above odds:



$$ 5/70 * 4/69 * 3/68 * 2/67 * 1/66 * 1/25 = 1 / 302575350 $$

                      Each fraction above represents the odds of getting at least one of your 5 balls and 1 megaball


## What are the chances of the Mega Millions to reach the billions again?

What are the forces at work causing this jackpot to get so large?  And what prevents it from reaching infinity? - The numbers of players.  As the jackpot increases in value and the Mega Millions receives news coverage and publicity, more and more players are attracted to the Mega Millions.  In fact it is quite possible that every single combination of tickets could be baught by customers, therefore gauranteeing that someone wins.  

For example we calculated above that the odds of winning are 1 in 302,575,350.  This also means that there is a total of 302,575,350 different combinations of winning jackpot numbers.  It is quite conceivable that Americans could buy this many tickets.

To get insight into our question above we need to get data on the total number of tickets sold at each cycle for the Mega Millions.  Luckily, the site http://www.lottoreport.com/ticketcomparison.htm records these numbers.  Let's take a glimpse below of the most recent numbers

```python
def reader(element1, element2):
    return(tr_elements[element1][element2].text_content())
```

```python
metadf = pd.read_excel("megametadata.xlsx")
columnnames= ["Date","megaplier #","5+1","5+1 Amount","5+0","4+1","4+0","3+1","3+0","2+1", "1+1","0+1","5+0+M","4+1+M","4+0+M","3+1+M","3+0+M","2+1+M", "1+1+M","0+1+M"]
df=pd.DataFrame(index=range(metadf["date"].count()),columns=columnnames)

i=0
for lottodate in metadf["date"]:
    
    url='http://www.megamillions.com/winning-numbers/'+str(lottodate.date().strftime('%#m-%#d-%Y'))
    #Create a handle, page, to handle the contents of the website
    page = requests.get(url)
    #Store the contents of the website under doc
    doc = lh.fromstring(page.content)
    #Parse data that are stored between <tr>..</tr> of HTML
    tr_elements = doc.xpath('//tr')
    
    df.iloc[i]=["12/05/02", "NA", reader(1,1),reader(1,2),reader(3,1),reader(4,1),reader(5,1),reader(6,1),reader(7,1),reader(8,1), reader(9,1),reader(10,1),reader(3,3),reader(4,3),reader(5,3),reader(6,3),reader(7,3),reader(8,3), reader(9,3),reader(10,3)]
    
    i+=1
    if i == metadf["date"].count():
        break
```

```python
df["megaplier #"]=metadf["megaplier"]
df["Date"]=metadf["date"]
df["ball1"]=metadf["ball1"]
df["ball2"]=metadf["ball2"]
df["ball3"]=metadf["ball3"]
df["ball4"]=metadf["ball4"]
df["ball5"]=metadf["ball5"]
df["megaball"]=metadf["megaball"]
df["totaltickets"]=metadf["totaltickets"]
```

```python
# Function to select years (columns) in databases

def keepyears(dataframe, startyear, endyear, yearstart=3):
    
    sliceleft=dataframe.drop(dataframe.columns[yearstart:dataframe.columns.get_loc(startyear)], axis=1)
    sliceright=sliceleft.drop(sliceleft.columns[sliceleft.columns.get_loc(endyear)+1:], axis=1)
    return(sliceright)
```

```python
# Removing all columns outside of start and end years

start="2000"
end="2016"

df1=keepyears(df1, start, end)
df2=keepyears(df2, start, end)
df3=keepyears(df3, start, end)
df4=keepyears(df4, start, end)
```

```python
# Function to select countries (rows) in database

def selectcountries(dataframe, countries):
    dataframe.index=dataframe["Country Name"]
    return(dataframe[dataframe["Country Name"].isin(countries)])
```

```python
# Removing all rows not included in country list

countryindex=["Burundi", "Ethiopia", "Kenya", "Rwanda", "Tanzania", "Uganda"]

df1=selectcountries(df1, countryindex)
df2=selectcountries(df2, countryindex)
df3=selectcountries(df3, countryindex)
df4=selectcountries(df4, countryindex)
```

```python
def plotdataframe2(*dataframes):
    
    plotsize=0
    for dataframe in dataframes:
        plotsize+=1
    
    plt.figure(figsize=(10, plotsize*3))
    
    plotcounts=0
    for dataframe in dataframes:
        plotcounts+=1
        
        plt.subplot(plotsize, 1, plotcounts)
        plt.plot(dataframe)
        plt.title("Graph "+str(plotcounts)+" - ")
        

    plt.tight_layout()
    plt.show()
               
plotdataframe2(df[["Date","totaltickets"]])
```

```python
def plotdataframe3(dataframe1, dataframe2):

    data1 = dataframe1.loc["Kenya",start:end]
    data2 = dataframe2.loc["Kenya",start:end]

    fig, ax1 = plt.subplots()
    fig.set_size_inches(10, 4)
    color = 'tab:red'
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel('exp', color=color)
    ax1.plot(data1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
    ax2.plot(data2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
    
    ax1.set_title("Kenya Tourism - Arrivals (Left) vs International Tourism (Right)")
    ax1.set_xlabel("year")
    ax1.set_ylabel("Total Arrivals")
    ax2.set_ylabel("Billions of $")
    
    plt.axvspan(7, 8, color='y', alpha=0.5, lw=0)
    plt.axvspan(11, 12, color='y', alpha=0.5, lw=0)
    
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()

        
plotdataframe3(df1,df4)
```
