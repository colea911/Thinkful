```python
# Random Distributions

Geometric = np.random.geometric(.5, 100)
Exponential = np.random.exponential(2, 100)
Poisson = np.random.poisson(2, 100)
Binomial = np.random.binomial(10,.5,100)
```

```python
# Graphing Random Distributions

plt.figure(figsize=(12, 3))


plt.subplot(1, 4, 2)
plt.hist(Poisson, color='c')
plt.title("Poisson")
plt.axvline(Poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.xlabel("Mean: "+str(round(Poisson.mean(),3))+" |  SD: "+str(round(Poisson.std(),3)))

plt.subplot(1, 4, 1)
plt.hist(Geometric, color='c')
plt.title("Geometric")
plt.axvline(Geometric.mean(), color='b', linestyle='solid', linewidth=2)
plt.xlabel("Mean: "+str(round(Geometric.mean(),3))+" |  SD: "+str(round(Geometric.std(),3)))

plt.subplot(1, 4, 3)
plt.hist(Exponential, color='c')
plt.title("Exponential")
plt.axvline(Exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.xlabel("Mean: "+str(round(Exponential.mean(),3))+" |  SD: "+str(round(Exponential.std(),3)))

plt.subplot(1, 4, 4)
plt.hist(Binomial, color='c')
plt.title("Binomial")
plt.axvline(Binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.xlabel("Mean: "+str(round(Binomial.mean(),3))+" |  SD: "+str(round(Binomial.std(),3)))

# Adds Spacing between subplots
plt.tight_layout()

plt.show()
```
