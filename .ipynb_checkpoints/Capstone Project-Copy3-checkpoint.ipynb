```python
# Function to select years (columns) in databases

def keepyears(dataframe, startyear, endyear, yearstart=3):
    
    sliceleft=dataframe.drop(dataframe.columns[yearstart:dataframe.columns.get_loc(startyear)], axis=1)
    sliceright=sliceleft.drop(sliceleft.columns[sliceleft.columns.get_loc(endyear)+1:], axis=1)
    return(sliceright)
```

```python
# Removing all columns outside of start and end years

start="2000"
end="2016"

df1=keepyears(df1, start, end)
df2=keepyears(df2, start, end)
df3=keepyears(df3, start, end)
df4=keepyears(df4, start, end)
```

```python
# Naming of Dataframes

df1.name=df1.iloc[1,2]
df2.name=df2.iloc[1,2]
df3.name=df3.iloc[1,2]
df4.name=df4.iloc[1,2]
```

```python
# Function to select countries (rows) in database

def selectcountries(dataframe, countries):
    dataframe.index=dataframe["Country Name"]
    return(dataframe[dataframe["Country Name"].isin(countries)])
```

```python
# Removing all rows not included in country list

countryindex=["Burundi", "Ethiopia", "Kenya", "Rwanda", "Tanzania", "Uganda"]

df1=selectcountries(df1, countryindex)
df2=selectcountries(df2, countryindex)
df3=selectcountries(df3, countryindex)
df4=selectcountries(df4, countryindex)
```

```python
# Calculate average tourists per year for each country

df1['mean'] = round(df1.mean(axis=1),-3)
df2['mean'] = round(df2.mean(axis=1),-3)
df3['mean'] = round(df3.mean(axis=1),-3)
df4['mean'] = round(df4.mean(axis=1),-3)

aggregates.index=df1.index
aggregates["Average Tourists"]=(df1["mean"])
aggregates["GDP Size"]=(df2["mean"])
aggregates["Population"]=(df3["mean"])
aggregates["Tourist Expenditures"]=(df4["mean"])

aggregates
```

```python
def plotdataframe2(*dataframes):
    
    plotsize=0
    for dataframe in dataframes:
        plotsize+=1
    
    plt.figure(figsize=(10, plotsize*3))
    
    plotcounts=0
    for dataframe in dataframes:
        plotcounts+=1
        
        plt.subplot(plotsize, 1, plotcounts)
        plt.title("Graph "+str(plotcounts)+" - "+dataframe.name)
        for country in countryindex:
            plt.plot(dataframe.loc[country,start:end])
        plt.legend(bbox_to_anchor=(1.02, 1.0), loc=2, borderaxespad=0)

    plt.tight_layout()
    plt.show()
               
plotdataframe2(df1,df2,df3,df4)
```

```python
def plotdataframe3(dataframe1, dataframe2):

    data1 = dataframe1.loc["Kenya",start:end]
    data2 = dataframe2.loc["Kenya",start:end]

    fig, ax1 = plt.subplots()
    fig.set_size_inches(10, 4)
    color = 'tab:red'
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel('exp', color=color)
    ax1.plot(data1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
    ax2.plot(data2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
    
    ax1.set_title("Kenya Tourism - Arrivals (Left) vs International Tourism (Right)")
    ax1.set_xlabel("year")
    ax1.set_ylabel("Total Arrivals")
    ax2.set_ylabel("Billions of $")
    
    plt.axvspan(7, 8, color='y', alpha=0.5, lw=0)
    plt.axvspan(11, 12, color='y', alpha=0.5, lw=0)
    
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()

        
plotdataframe3(df1,df4)
```

```python
# Import Terrorism Data 

dfterr = pd.read_excel("terrorismdata2.xlsx")

```

```python
# Remove all rows in which data is not one of our countries of interest


dfterr=dfterr[dfterr["country_txt"].isin(countryindex)]
```

```python
# Filter dataset by columns

dfterr=dfterr.filter(["iyear","country_txt","nkill"])
```

```python
# Group by year and country 

terroristdatacount=dfterr.groupby(["country_txt","iyear"])[["iyear"]].count()
terroristdatamean=dfterr.groupby(["country_txt","iyear"])[["nkill"]].mean()

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
for country in countryindex:
    plt.plot(terroristdatacount.loc[country,:],label=country)
plt.title("Terrorist Incidents by Country")
plt.legend(bbox_to_anchor=(1.02, 1.0), loc=2, borderaxespad=0)

plt.subplot(2, 1, 2)
for country in countryindex:
    plt.plot(terroristdatamean.loc[country,:],label=country)
plt.title("Terrorist Incidents by Country")
plt.legend(bbox_to_anchor=(1.02, 1.0), loc=2, borderaxespad=0)

plt.tight_layout()
plt.show()


    
```

```python
def plotdataframe(dataframe1, dataframe2):

    data3 = terroristdatacount.loc["Kenya",:]
    data4 = terroristdatamean.loc["Kenya",:]

    fig, ax1 = plt.subplots()
    fig.set_size_inches(10, 4)
    color = 'tab:red'
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel('exp', color=color)
    ax1.plot(data3, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
    ax2.plot(data4, color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    ax1.set_xlim(2000, 2016)
    
    
    ax1.set_title("Incidents of Terrorism vs Casualities of Terrorism")
    ax1.set_xlabel("year")
    ax1.set_ylabel("Incidents")
    ax2.set_ylabel("Casualties")
    
    plt.axvspan(7, 8, color='y', alpha=0.5, lw=0)
    plt.axvspan(11, 12, color='y', alpha=0.5, lw=0)
    
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()

        
plotdataframe(df1,df4)
```
