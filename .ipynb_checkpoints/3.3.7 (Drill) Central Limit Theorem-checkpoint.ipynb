```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
pop1 = np.random.binomial(10,.2,10000)
pop2 = np.random.binomial(19,.5,100)

plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.axvline(pop1.mean(), color='b', linestyle='solid', linewidth=2)
plt.hist(pop1, color='c')
plt.title("Pop1")
plt.xlabel("Mean: "+str(round(pop1.mean(),3))+" |  SD: "+str(round(pop1.std(),3)))

plt.subplot(1, 2, 2)

plt.axvline(pop2.mean(), color='b', linestyle='solid', linewidth=2)
plt.hist(pop2, color='c')
plt.title("pop2")
plt.xlabel("Mean: "+str(round(pop2.mean(),3))+" |  SD: "+str(round(pop2.std(),3)))

plt.tight_layout()
plt.show()
```

```python
##### sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

sample2 = np.random.choice(pop2, 100, replace=True)

plt.figure(figsize=(10, 3))

plt.subplot(1, 3, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.title("Sample 100")
plt.xlabel("Mean1: "+str(round(sample1.mean(),3))+" |  Mean2: "+str(round(sample2.mean(),3)))

sample3 = np.random.choice(pop1, 1000, replace=True)
sample4 = np.random.choice(pop2, 1000, replace=True)

plt.subplot(1, 3, 3)
plt.hist(sample3, alpha=0.5, label='sample 1') 
plt.hist(sample4, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right')
plt.title("Sample 1000")
plt.xlabel("Mean1: "+str(round(sample3.mean(),3))+" |  Mean2: "+str(round(sample4.mean(),3)))

sample5 = np.random.choice(pop1, 20, replace=True)
sample6 = np.random.choice(pop2, 20, replace=True)

plt.subplot(1, 3, 1)
plt.hist(sample5, alpha=0.5, label='sample 1') 
plt.hist(sample6, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.title("Sample 20")
plt.xlabel("Mean1: "+str(round(sample5.mean(),3))+" |  Mean2: "+str(round(sample6.mean(),3)))

plt.tight_layout()
plt.show()


print("Sample 20 Graph: Mean = "+str(sample5.mean())+" and SD = "+str(sample6.std()))
print("Sample 200 Graph: Mean = "+str(sample1.mean())+" and SD = "+str(sample2.std()))
print("Sample 1000 Graph: Mean = "+str(sample3.mean())+" and SD = "+str(sample4.std()))
```
